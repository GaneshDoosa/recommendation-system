```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns',200)
pd.set_option('display.max_rows',200)
from sklearn.metrics.pairwise import cosine_similarity
```

```python
ratings = pd.read_csv('F:\\RecommendationEngine\\dataset\\ml-latest-small\\ratings.csv')
print('Ratings Dataset Shape:', ratings.shape)
print('Unique Users:', ratings.userId.nunique())
ratings.head()
```

```python
movies = pd.read_csv('F:\\RecommendationEngine\\dataset\\ml-latest-small\\movies.csv')
print(movies.shape)
movies.head()
```

```python
merged = ratings.merge(movies,left_on='movieId',right_on='movieId',sort=True)
merged.head()
```

```python
merged = merged[['userId','title','rating']]
movieratings = merged.pivot_table(index=['userId'],columns=['title'],values='rating')
movieratings.replace({np.nan:0},inplace=True)
print(movieratings.shape)
movieratings.head()
```

```python
user_similarity = cosine_similarity(movieratings)
print(type(user_similarity))
print('Shape of User Similarity Matrix', user_similarity.shape)
user_similarity[0]
```

```python
user_sim_df = pd.DataFrame(user_similarity,index=movieratings.index,columns=movieratings.index)
print(user_sim_df.shape)
print(user_sim_df.head())
```
